# ABIDE

```python
import matplotlib.pyplot as plt
import numpy as np

# Read the data from the file
file_path = './ABIDE_density.txt'

with open(file_path, 'r') as file:
    lines = file.readlines()

# Initialize data structures
data = {}
current_model = None

# Parse the data
for line in lines:
    line = line.strip()
    if line.startswith('['):
        current_model = line.strip('[]')
        data[current_model] = {'density': [0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.5, 0.7, 1],
                               'auc_mean': [], 'auc_std': [],
                               'acc_mean': [], 'acc_std': [],
                               'sens_mean': [], 'sens_std': [],
                               'spec_mean': [], 'spec_std': []}
    elif line:
        parts = line.split(',')
        auc = parts[0].split('=')[1].split('+-')
        acc = parts[1].split('=')[1].split('+-')
        sens = parts[2].split('=')[1].split('+-')
        spec = parts[3].split('=')[1].split('+-')

        
        data[current_model]['auc_mean'].append(float(auc[0]))
        data[current_model]['auc_std'].append(float(auc[1]))
        data[current_model]['acc_mean'].append(float(acc[0]))
        data[current_model]['acc_std'].append(float(acc[1]))
        data[current_model]['sens_mean'].append(float(sens[0]))
        data[current_model]['sens_std'].append(float(sens[1]))
        data[current_model]['spec_mean'].append(float(spec[0]))
        data[current_model]['spec_std'].append(float(spec[1]))



import matplotlib.pyplot as plt
import numpy as np

def plot_dataset(data):
    markers = ['v', '<', 'o', 's', 'd', '^', 'p', '*', 'p', '*']
    linestyles = ['-', '-', '-', '-', '-', '-', '-', '-', ':', ':']
    colors = ['grey', 'orange', 'b', 'm', 'r', 'g', 'c', 'k', 'c', 'k']

    linewidth = 3  # Set the linewidth for all plots
    markersize = 13  # Set the marker size for all plots

    # Function to plot data
    def plot_with_shadow(ax, x, y_mean, y_std, label, marker, linestyle, color):
        ax.plot(x, y_mean, label=label, marker=marker, linestyle=linestyle, color=color, linewidth=linewidth, markersize=markersize)
        ax.grid(False)

    # Plotting
    fig, axs = plt.subplots(1, 2, figsize=(16, 6))
    ax1, ax2 = axs[0], axs[1]

    for i, (model, values) in enumerate(data.items()):
        if i <= 5:
            plot_with_shadow(ax1, values['density'], values['auc_mean'], values['auc_std'], model, markers[i], linestyles[i], colors[i])
        else:
            plot_with_shadow(ax2, values['density'], values['auc_mean'], values['auc_std'], model, markers[i], linestyles[i], colors[i])

    font_size = 27

    ax1.set_xlabel('Graph Density', fontsize=font_size)
    ax1.set_ylabel('AUROC', fontsize=font_size)
    ax1.tick_params(axis='both', which='major', labelsize=font_size, width=2, length=10, direction='inout')
    ax1.set_ylim(top=0.76)
    ax1.grid(True, which='major', axis='y', linestyle='--', linewidth=1)
    ax1.grid(True, which='major', axis='x', linestyle='--', linewidth=1)


    ax2.set_xlabel('Graph Density', fontsize=font_size)
    ax2.tick_params(axis='both', which='major', labelsize=font_size, width=2, length=10, direction='inout')
    ax2.set_ylim(top=0.76)
    ax2.grid(True, which='major', axis='y', linestyle='--', linewidth=1)
    ax2.grid(True, which='major', axis='x', linestyle='--', linewidth=1)


    plt.tight_layout()

    plt.savefig(f'./ABIDE_density.pdf', format='pdf', bbox_inches='tight')
    plt.show()
    plt.close()

plot_dataset(data)  

```
